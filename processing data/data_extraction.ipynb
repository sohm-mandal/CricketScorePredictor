```python
import numpy as np
import pandas as pd
import yaml
from yaml import safe_load
import os
from tqdm import tqdm
```

```python
path = r'C:\Users\soham\OneDrive\Desktop\New folder (4)\dataset\t20s'
filenames = []
for file in os.listdir(path):
    filenames.append(os.path.join(path, file))
```

```python
filenames[0:10]
```

```python
df_list = []
cnt = 1

for file in tqdm(filenames):
    with open(file, 'r') as f:
        data = yaml.safe_load(f)
        if isinstance(data, dict):
            data = [data]  # Ensure the data is a list of dictionaries
        try:
            df = pd.json_normalize(data)
            df['match_id'] = cnt
            df_list.append(df)
            cnt += 1
        except Exception as e:
            print(f"Failed to normalize JSON data from {file}: {e}")

final_df = pd.concat(df_list, ignore_index=True)
```

```python
cnt = 1
delivery_df_list = []  
for index, row in matches.iterrows():
    cnt += 1
    ball_of_match = []
    batsman = []
    bowler = []
    runs = []
    player_of_dismissed = []
    teams = []
    batting_team = []
    bowling_team = []
    match_id = []
    city = []
    venue = []
    
    # taking batting and bowling teams
    team_1 = row['innings'][0]['1st innings']['team']
    
    if len(row['innings']) > 1:
        team_2 = row['innings'][1]['2nd innings']['team'] 
    else:
        team_2 = row['innings'][0]['1st innings']['team']
    
    # taking data for all
    for ball in row['innings'][0]['1st innings']['deliveries']:
        for key in ball.keys():
            match_id.append(cnt)
            batting_team.append(team_1)
            bowling_team.append(team_2)
            teams.append(row['info.teams'])
            ball_of_match.append(key)
            batsman.append(ball[key]['batsman'])
            bowler.append(ball[key]['bowler'])
            runs.append(ball[key]['runs']['total'])
            city.append(row['info.city'])
            venue.append(row['info.venue'])
            try:
                player_of_dismissed.append(ball[key]['wicket']['player_out'])
            except KeyError:
                player_of_dismissed.append('0')
                
    newrow = pd.DataFrame({
        'match_id': match_id,
        'teams': teams,
        'batting_team': batting_team,
        'bowling_team': bowling_team,
        'ball': ball_of_match,
        'batsman': batsman,
        'bowler': bowler,
        'runs': runs,
        'player_dismissed': player_of_dismissed,
        'city': city,
        'venue': venue
    })
    
    delivery_df_list.append(newrow) 

delivery_df = pd.concat(delivery_df_list, ignore_index=True)

delivery_df
```

```python
# i want top 15 teams according to number of matches
# sorted in descending order
batting_team_counts = delivery_df['batting_team'].value_counts()
teams = []
# sorted in descending order
for country, count in batting_team_counts.items():
    teams.append(country)
    
teams = teams[0:15]
```

```python
delivery_df = delivery_df[delivery_df['batting_team'].isin(teams)]
delivery_df = delivery_df[delivery_df['bowling_team'].isin(teams)]
```
